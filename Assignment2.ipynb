2.Explore:
In this part, the appearance of data and the relationship between different data will be explored in the form of data visualization. 

Firstly, I want to follow the last step about some of summary the statistic

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline


tem1 = DataMergeFinal.value_counts('Species')
#tem1 is a temporary variable used to temporarily store the extracted series of each number of specifications

SpeciesAndNumber = pd.DataFrame({'Species':tem1.index, 'Num':tem1.values})
plt.figure(figsize=[10,20])#canvas size


SpeciesBar = sns.barplot(y='Species', x='Num', data=SpeciesAndNumber)

#display the value of every bar
for container in SpeciesBar.containers:
    SpeciesBar.bar_label(container)
```

We can more intuitively see the number of each tree through this visualization. The largest number is Black locust, which has 177 trees. The second largest number is European hackberry, which is 150 trees. These two kinds of trees account for more than half of the sampling.

Whether the largest number of trees are also the most widely distributed and appear on each plot, so next try to find the relationship between tree species distribution and plot

```python
# plot
from matplotlib.pyplot import MultipleLocator
plt.figure(figsize=[10,20])
sns.violinplot(x='Plot', y='Species', data=DataMergeFinal, kind='violin',scale='width', cut=0 )
#cut = 0 allows the violin graph to terminate at the extreme value without depicting the entire distribution


#Let the scale unit of the X axis be 1
ax = plt.gca()
ax.xaxis.set_major_locator(MultipleLocator(1))
```

In this violin chart, the European hackberry, second lagrest one, is indeed distributed on every 'plot'. But the largest, the Black locust did not showed in 'the Plot 1'.

Other things we can receive from the chart include that the Field elm is mainly concentrated after plot 16, and the European black elderberry is mainly concentrated between plot 5 to plot 13.

The Common fig just be showed between the plot 16 to 22.

Next, let's look at the height distribution of all trees

```python

plt.figure(figsize=[15,8])
sns.violinplot(x='Total Height (m)',  data=DataMergeFinal, kind='violin',scale='width', cut=0 )
ax = plt.gca()
ax.xaxis.set_major_locator(MultipleLocator(1))
```

We can see that the approximate average tree height is between 9m and 10m, but the most distributed trees are between 6m and 8m. The number of trees with a height of 12 to 20 meters gradually decreases smoothly

Can we make a inference that there are some tree species occupying a large number, and many trees in these species may be concentrated between 6-8 meters in height.. Therefore, the relationship between tree species and tree height distribution can be found through visualization


```python
plt.figure(figsize=[10,20])
sns.violinplot(x='Total Height (m)', y='Species',  data=DataMergeFinal, kind='violin',scale='count', cut=0 )
ax = plt.gca()
ax.xaxis.set_major_locator(MultipleLocator(1))
```

In this violin chart, the total height and the species are linked together and we can see that height of the second largest number tree.

European hackberry, is mainly concentrated between 4m to 10m. The Field elm and the Sweet cherry have the similar situation, which are randed in the 4 and 6 about the trees'number.

Of course, the two kinds of trees with low overall height mentioned above, the European hackberry and the Field elm, also provide very high value and close to very high value of tree height, but the number is small


After the tree height, I want to look at another data about the tree size, DBH


In fact, DBH should be measured several times at different heights of trees in order to accurately measure the growth of a tree. This is also the meaning of multiple groups of DBH and corresponding height in the original data. However, since most of them are empty columns in the original table, only DBH 1 is used here to represent the original meaning to a certain extent


```python
plt.figure(figsize=[15,8])
sns.violinplot(x='DBH 1 (cm)',  data=DataMergeFinal, kind='violin',scale='width', cut=0 )
plt.show()
```

We can see from the above distribution diagram that the DBH of the tree is mainly distributed in the part below 25 cm.But there are trees with very large extremum

From the current data, to more accurately observe the growth of trees, we need to judge the overall growth trend of trees in combination with tree height. Will the higher the trees, the greater their DBH?

```python
sns.lmplot( x="Total Height (m)", y="DBH 1 (cm)", data = DataMergeFinal, fit_reg=True)

plt.show()
```

Generally speaking, the tree height and DBH follow the linear relationship of positive correlation to a certain extent, that is, the higher the tree is, the larger its DBH is, but there are many points of principle regression line. It may be that different trees have different ratio relationships or growth conditions, which leads to the fact that DBH growth is much faster or slower than other trees, under the same tree height

Therefore, the factors of tree species are introduced and divided into different diagrams to observe the differences in tree height and DBH growth rate among different tree species

```python
sns.lmplot( x="Total Height (m)", y="DBH 1 (cm)", data = DataMergeFinal, hue= "Species",col = "Species", col_wrap=5, height=4, fit_reg=True)

plt.show()
```

From the scatter diagram above, we can see that the slope of the regression line formed by the tree height and DBH of some trees is much smaller than that of other tree species, which shows that these trees do not become thicker with the increase of tree height.

Like Sweet Cherry and European blackelderberry.

Field elm  and White mulberry pull the slope of the whole regression line very high because of several special points. To some extent, such a slope does not mean that it can grow coarser at the same tree height. It is still necessary to observe the main aggregation of scattered points.

Robinia spp is the only tree species with a downward trend in tree height and DBH. However, its small number does not mean that this kind of tree will become thinner with growth

The largest number of tree species, Black locust, is the closest to the regression of all scattered points. It is speculated that all trees grow roughly according to this regular.


Crown condition is another indicator used to describe the growth of trees. Here we can see whether all trees have complete crown development and how many trees have different crown development


```python
tem2 = DataMergeFinal.value_counts('Crown: Condition')
SpeciesAndNumber = pd.DataFrame({'Crown: Condition':tem2.index, 'Num':tem2.values})
plt.figure(figsize=[10,20])


CrownBar = sns.barplot(y='Crown: Condition', x='Num', data=SpeciesAndNumber)


for container in CrownBar.containers:
     CrownBar.bar_label(container)
```

We can see that the canopy of most trees has grown to more than 70%. Next, you might as well pay attention to which plots have lower values.


```python
plt.figure(figsize=[10,20])
sns.violinplot(x='Plot', y='Crown: Condition',  data=DataMergeFinal, kind='violin',scale='count', cut=0 )
ax = plt.gca()
ax.xaxis.set_major_locator(MultipleLocator(1))
```

It can be seen that many trees with low crown percentage are also distributed in many different plots, which means that different growth conditions may not be caused by different locations, but may be caused by differences between tree

There is another data in the table, but I don't quite understand its meaning. I guess it may be the number of trees or the ratio of crown shadow area to the area of the land where it is located, which is used to represent the shadow coverage of trees in this area


```python
tem3 = DataMergeFinal.value_counts('% Tree')
SpeciesAndNumber = pd.DataFrame({'% Tree':tem3.index, 'Num':tem3.values})
plt.figure(figsize=[10,20])

SpeciesBar = sns.barplot(y='% Tree', x='Num', data=SpeciesAndNumber)

for container in SpeciesBar.containers:
    SpeciesBar.bar_label(container)
```

You can see that there are still a lot of plots covered by the shade or the main area covered by the shade. These plots may be a good place to avoid the sun in the whole park
